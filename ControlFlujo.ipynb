# CONTROL DE FLUJO 

## EXPRESIONES CONDICIONALES: if, if-else, if-elif-else.

En la mayoría de las ocasiones los programas no se ejecutan siguiendo todas y cada una de las lineas de código, o sea que el flujo de las instrucciones puede variar, y eso sucede gracias a determinadas estructuras de control que son clave y que nos encontraremos muy a menudo en cualquier programa o script. Python tiene tres tipos de expresiones para controlar el flujo de ejecución y una de ellas es la instrucción condicional if.

## if simple

```python
a=8
if a>3:
    print ("a es mayor que 3")
```

Las instrucciones que aparecen despues de la sentencia de control forman un bloque de código que debe de estar indentado para indicar que pertenece a la sentencia if. Esto significa que todo este bloque se ejecutará si la **sentencia if** es verdadera. si el falsa no entrará en el bloque de intrucciones.

```python
a=4
if a<8:
    print ("a es mayor que 8")

print ("Salida del programa")
```

También podemos utilizar varias sentencias **if anidadas**.

```python
a=40
if a>30:
    if a<50:
        print ("a es mayor que 30 y menor que 50")
```

Un resultado similar conseguimos con el operador lógico **and** para evaluar dos a más expresiones condicionales.

```python
a=40
if a>30 and a<50:
        print ("a es mayor que 30 y menor que 50")
```

## if-else

**Else** podría tradicirse como **"sino"** o **"en caso contrario"**. Esto nos va a servir para ejecutar un bloque de código alternativo en el caso de que la condición que sigue a if sea falsa y ese caso se ejecutará la instrucción que sigue a else.**

 <span style="color:red">**Función input()**</span>

<span style="color:red">Como podemos ver esta función me permite recibir una información del programa y luego me permite responder introduciendo un dato que me devuelve. Pero como podemos ver dicho dato es por defecto una cadena, por eso la vemos entre comillas, cuando en realidad nosotros queríamos introducir un dato numérico para poder hacer después operaciones con el. Para resolver esto utilizamos otra función, **eval** que nos va a evaluar la función input para devolvernos el tipo de dato que queremos.</span>

```python
eval(input("Introduce un número: "))
```

```python
num=int(input("Introduce un número: "))
if num%2==0:
    print ("El número que has introducido es par.")
else:
    print ("El número que has introducido es impar.")
```

## if-elif-else

Elif significa **else if** podría traducirse por **en caso contrario si...** y con ella podremos utilizar varias condiciones alternativas. En el momento que alguna sea verdadera se sale de la expresión condicional.

```python
num=int(input("Introduce un número: "))
if num>0:
    print("El número introducido es positivo")
elif num==0:
    print ("El número introducido es 0")
else:
    print("El número introducido es negativo")
```

## EXPRESIONES ITERATIVAS: bucle for, bucle while

### EL BUCLE for ###

En ocasiones debemos repetir varias veces un mismo bloque de código para poder obtener el resultado deseado. Esto lo podemos realizar mediante la instrucción for donde el bucle es controlado por una variable que recorre una serie de valores. A medida que la variable va pasando por cada uno de los valores se va ejecutando en cada paso el bloque de código que sigue. A cada una de estas repeticiones se le llama **iteración**.

```python
for letra in "actgtg":
    print (letra)
```

```python
for i in [1, "aatgac", 3, "MPTAGAN"]:
    print (i)
```

Otra cosa que podemos hacer con un bucle for es modificar rápidamente los elementos de una lista,al mismo tiempo que iteramos sobre ella.

```python
lista=[5,6,7,8]
indice=0
for elemento in lista:
    lista[indice]*=2
    indice+=1
    #indice=indice+1
lista
```

La función **enumerate()** nos devuelve el índice  y valor del objeto iterable.

```python
lista=[6,7,8,9]
for indice,numero in enumerate (lista):
    lista[indice]*=2
    print(indice)
lista
```

 <span style="color:red">**Compresiones de listas**</span>

 <span style="color:red">Las compresiones de listas ofrecen una manera concisa de crear listas. Su uso más común es para crear nuevas listas donde cada elemento es el resultado de algunas operaciones aplicadas a cada miembro de otra secuencia o iterable.

```python
listNum=[1,2,3,4,5]
listCuadrados=[numero**2 for numero in listNum]
print (listCuadrados)
```

### EL BUCLE while ###

El bucle while es similar al bucle for, pero en este caso no se pruduce una iteración sobre un objeto y por tanto no termina cuando el objeto es recorrido, sino que la iteración termina cuando una condición lógica no se cumple.

```python
numero=0
while numero <=10:
    print ("hola")
    numero+=1
    #numero=numero+1
```

En muchas ocasines podremos hacer lo mismo con for que con while, pero si no conocemos de antemano el número de iteraciones nos resultaría imposible hacerlo con for, por ejemplo en este caso:

```python
suma=0
numero=1
while numero!=0:
    numero=eval(input("Introduce un numero: "))
    suma=suma+numero
print ("La suma de todos los numeros es",suma)
```

Dos instrucciones importantes que debemos conocer a la hora de utilizar los bucles son **break** y **continue**. Veamos el siguiente código.

```python
num=0
while num>=0:
    num+=1
    print ("hola")
    if num>7:
        break
```

Cuando se ejecuta la instrucción **break** automáticamente se sale del bucle.

```python
num=0
while num<10:
    num+=1
    if num==7:
        continue
    print (num)
```

La instrucción **continue** va a saltarse la iteración que le indiquemos, pero el blucle se va a continuar ejecutando.
