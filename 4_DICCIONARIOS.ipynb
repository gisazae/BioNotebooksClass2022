# TIPOS DE DATOS EN PYTHON 

## DICCIONARIOS (dict)
Un diccionario es un objeto de la clase dict, que contiene pares no ordenados, clave & valor ... 

### 1)  Para crear un diccionario ...

```python
#las claves no pueden estar repetidas 
d={"seq1":"atgctgat...", "seq2":"tgctgatc...", "seq3":"tgctgat..."}
d
len(d)
```

### 2) Operadores para diccionarios

**<span style="color:red">Operador [clave]</span> : índice. Devuelve el valor de la clave indicada.**

```python
d["seq1"]
```

**<span style="color:red">Operador in</span> : en (dentro de). Devuelve un dato de tipo booleano (True, False).**

```python
"seq2" in d
```

**<span style="color:red">Operador not in</span> : no en (no dentro de). Devuelve un dato de tipo booleano.**

```python
"seq3" not in d
```

**<span style="color:red">Operador del</span> : Elimina la entrada indicada con la clave.**

```python
del d["seq3"]
d
```

### 3) Algunas funciones predeterminadas

**<span style="color:red">Función len()</span>: Devuelve el número de entradas del diccionario.**

```python
len(d)
```

**<span style="color:red">Función sum()</span>: Devuelve la suma de las claves del diccionario.**

```python
d2={1:"Tyrel", 2:"Stark", 3:"Lannister", 4:"Targaryen"}
sum(d2)
```

**<span style="color:red">Función max()</span>: Devuelve el valor máximo de las claves del diccionario.**

```python
max(d2)
```

**<span style="color:red">Función min()</span>: Devuelve el valor mínimo de las claves de diccionario.**

```python
min(d2)
```

### 4) Métodos de la clase dict

**<span style="color:red">Método get(clave)</span>: Devuelve el valor de la clave.**

```python
d={"seq1":"atgctgat...", "seq2":"tgctgatc...", "seq3":"tgctgat..."}
d.get("seq3")
```

**<span style="color:red">Método keys()</span>: Devuelve las claves del diccionario.**

```python
d.keys()
```

**<span style="color:red">Método values()</span>: Devuelve los valores del diccionario.**

```python
d.values()
```

**<span style="color:red">Método items()</span>: Devuelve las entradas del diccionario.**

```python
d.items()
```

**<span style="color:red">Método pop(clave)</span>: Devuelve el valor de la entrada eliminada.**

```python
d.pop("seq1")
```

**<span style="color:red">Método clear()</span>: Elimina todas las entradas del diccionario.**

```python
d.clear()
d
```
