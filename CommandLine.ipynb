### 6. Variables de entorno

    > export NORMALIZATION="divs"
    > nscore tmp.csv | head
    
    > unset NORMALIZATION
    > nscore tmp.csv > tmp-mult.csv
    > export NORMALIZATION="div"
    > nscore tmp.csv > tmp-div.csv
