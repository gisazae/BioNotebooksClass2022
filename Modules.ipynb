Para utilizar el módulo en nuestro programa utilizamos la palabra **import**. 

```python
import test1
```

Nuestro scritp a parte de ejecutar una acción puede contener también una o varias **funciones** que podemos importar y utilizar en nuestro programa. **Si importamos todo el módulo** para utilizar alguna función debemos escribir el **nombre del módulo + operador punto + nombre de la función(paramétros)**, como en el siguiente ejemplo:

```python
import test2
res=test2.sumar(3,6)
res
```

**Si importamos solo la función que necesitamos hace falta escribir el nombre del módulo para acceder a la función.**

```python
from test2 import sumar
sumar(5,4)
```

## MÓDULO os

**El <span style="color:red">módulo os</span> nos permite acceder a funcionalidades dependientes del Sistema Operativo, sobre todo, aquellas que nos refieren información sobre el entorno del mismo y nos permiten manipular la estructura de directorios.**


Para conocer el directorio actual:

```python
import os
os.getcwd()
```

Para subir en el directorio:

```python
os.chdir("..")
```

```python
os.getcwd()
```

Para listar el directorio actual:

```python
os.listdir()
```

Para cambiar a otro directorio:

```python
os.chdir("FUNCIONES")
```

```python
os.listdir()
```

Para borrar un archivo.

```python
os.remove("archivo")
```

Para crear un directorio.

```python
os.mkdir("dir")
```

Para borrar un directorio vacio.

```python
os.rmdir("dir")
```

Para renombrar un archivo o directorio.

```python
os.rename("seqs", "seqs2")
```

Para saber si existe un archivo o carpeta.

```python
os.path.exists("seqs2")
```

Para saber si es archivo o carpeta.

```python
os.path.isfile("fichero_salida2.fasta")
```
