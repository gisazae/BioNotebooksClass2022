### 2) Operadores de listas

**<span style="color:red">Operador +</span> : Suma. Concatena listas.**

```python
lista1=[1,2,3,4]
lista2=[5,6,7,8,9]
lista1+lista2
```

**<span style="color:red">Operador *</span> : Multiplicación. Repite listas.**

```python
4*lista1
```

**<span style="color:red">Operador ==</span> : Igual que. Devuelve un dato de tipo booleano (True, False).**

```python
lista1==lista2
```

**<span style="color:red">Operador !=</span> : Distinto de. Devuelve un dato de tipo booleano (True, False).**

```python
lista1!=lista2
```

**<span style="color:red">Operador in</span> : en (dentro de). Devuelve un dato de tipo booleano (True, False).**

```python
8 in lista2
```

**<span style="color:red">Operador not in</span> : no en (no dentro de). Devuelve un dato de tipo booleano.**

```python
3 not in lista1
```

**<span style="color:red">Operador [ ]</span> : índice. Devuelve el item que ocupa la posición indicada.**

**En Python se empieza a contar desde el carácter 0.**

```python
lista1[2]
```

```python
lista1[-1]
```

**Si queremos obtener una sublista podemos utilizar intervalos, separados por dos puntos ":". El procedimiento es el mismo que en las cadenas.**

```python
lista3=[1,2,3,4,5,6,7,8,9]
lista3[2:8]
```

**Las listas son objetos mutables.**

```python
lista3[2]=12
lista3
```

### 3) Algunas funciones predeterminadas

**<span style="color:red">Función len()</span>: Devuelve el número de elementos de la lista.**

```python
lista4=[10,20,30,40,50]
len(lista4)
```

**<span style="color:red">Función sum()</span>: Devuelve la suma de los elementos de la lista.**

```python
sum(lista4)
```

**<span style="color:red">Función max()</span>: Devuelve el valor máximo de la lista.**

```python
max(lista4)
```

**<span style="color:red">Función min()</span>: Devuelve el valor mínimo de la lista.**

```python
min(lista4)
```

**<span style="color:red">Función print()</span>: Imprime el resultado por pantalla.**

```python
print (lista4)
```

### 4) Métodos de la clase list ###

**<span style="color:red">Método append(objeto)</span>: Añade un elemento al final de la lista.**

```python
lista5=["a","b","c","d","e"]
lista5.append("f")
lista5
```

**<span style="color:red">Método extend(lista)</span>: Añade otra lista al final de la inicial.**

```python
lista5.extend(["g","h"])
lista5
```

**<span style="color:red">Método insert(entero, objeto)</span>: Añade un elemento en una posición determinada.**

```python
lista5.insert(3, "hola")
lista5
```

**<span style="color:red">Método count(objeto)</span>: Devuelve el número de apariciones del elemento indicado.**

```python
lista5.count("a")
```

**<span style="color:red">Método pop(entero)</span>: Elimina el objeto indicado por el índice. Devuelve el objeto eliminado. Si no se indica nada elimina y devuelve el último objeto de la lista.**

```python
lista5.pop(0)
```

```python
lista5
```

**<span style="color:red">Método remove(objeto)</span>: Elimina de la lista la primera aparición del objeto indicado.**

```python
lista5.remove("b")
lista5
```

**<span style="color:red">Método reverse()</span>: Invierte los elementos de la lista.**

```python
lista5.reverse()
lista5
```

**Vamos a ver algunos metodos que podemos usar con la clase list()**

```python
seq1="atgctgatg"
seq2="atctcgtgatga"
seq3="tgatctgt"
lista_seq=[seq1, seq2, seq3]
print(lista_seq)
```

```python
#si quisieramos añadir una nueva secuencia a nuestra lista

seq4="tgtctatgcta"
lista_seq.append(seq4)
print(lista_seq)
```

```python
#si se incluye mas de un elemento a la lista usamos extend
seq5="tgtgtcgatcggg"
seq6="gtgggccaatt"
lista_seq2=[seq5, seq6]
lista_seq.extend(lista_seq2)
print(lista_seq)
```

```python
#por su parte, si quisieramos eliminar el último elemento de la lista

lista_seq.pop()
print(lista_seq)
```

```python
#Eliminando un objeto especifico

lista_seq.pop(2)
lista_seq
```

```python
#para remover una porción que ya habiamos incluido anteriormente 

lista_seq.remove(seq1)
print(lista_seq)
```
