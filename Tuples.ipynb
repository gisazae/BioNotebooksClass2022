# TIPOS DE DATOS EN PYTHON 

## TUPLAS (tuple)

Es un objeto similar a las listas, son objetos inmutables, a diferencia de las listas. Nos confiere seguridad extra, puesto que no se podrían modificar accidentalmente

### 1)  Para crear una tupla ...

```python
t=(1,"actg",3.41, [1,2,3,4])
t
```

### 2) Operadores de tuplas

**<span style="color:red">Operador +</span> : Suma. Concatena tuplas.**

```python
t1=[1,2,3,4]
t2=[5,6,7,8,9]
t1+t2
```

**<span style="color:red">Operador *</span> : Multiplicación. Repite tuplas.**

```python
4*t1
```

**<span style="color:red">Operador ==</span> : Igual que. Devuelve un dato de tipo booleano (True, False).**

```python
t1==t2
```

**<span style="color:red">Operador !=</span> : Distinto de. Devuelve un dato de tipo booleano (True, False).**

```python
t1!=t2
```

**<span style="color:red">Operador in</span> : en (dentro de). Devuelve un dato de tipo booleano (True, False).**

```python
2 in t1
```

**<span style="color:red">Operador not in</span> : no en (no dentro de). Devuelve un dato de tipo booleano.**

```python
3 not in t1
```

**<span style="color:red">Operador [ ]</span> : índice. Devuelve el item que ocupa la posición indicada.**

**En Python se empieza a contar desde el carácter 0.**

```python
t1[2]
```

**Si queremos obtener una sublista podemos utilizar intervalos, separados por dos puntos ":". El procedimiento es el mismo que en las cadenas y en las listas.**

```python
t3=[1,2,3,4,5,6,7,8,9]
t3[2:8]
```

### 3) Algunas funciones predeterminadas

**<span style="color:red">Función len()</span>: Devuelve el número de elementos de la tupla.**

```python
t4=(34, 54, 7, -3)
len(t4)
```

**<span style="color:red">Función sum()</span>: Devuelve la suma de los elementos de la tupla.**

```python
sum(t4)
```

**<span style="color:red">Función max()</span>: Devuelve el valor máximo de la tupla.**

```python
max(t4)
```

**<span style="color:red">Función min()</span>: Devuelve el valor mínimo de la lista.**

```python
t3=[19,22,33,44,5,6,7,8,9]
min(t3)
```
