# TIPOS DE DATOS EN PYTHON 

## CONJUNTOS (set)
Son objetos de la clase SET, son mutables, almacenan diferentes tipos de objetos pero no de forma ordenada, los objetos no pueden estar repetidos entre si. 

Una de las ventajas es que podemos buscar un elemento entre grandes cantidades de datos o de colecciones muy extensas, esto de manera muy rápida y eficiente. 

### 1)  Para crear un conjunto ...

```python
c={534, "actg", 5.7, (1,2,3,4)}
c
```

### 2) Operadores para conjuntos

**<span style="color:red">Operador |</span> : Unión. Devuelve un conjunto con la unión de dos conjuntos.**

```python
c1={1,2,3,4,5,6}
c2={5,6,7,8,9}
c1|c2
```

**<span style="color:red">Operador &</span> : Intersección. Devuelve un conjunto con la intersección de dos conjuntos.**

```python
c1&c2
```

**<span style="color:red">Operador -</span> : Diferencia. Devuelve un conjunto con los elementos que están en c1 pero no en c2.**

```python
c1-c2
```

**<span style="color:red">Operador ^</span> : Diferencia simétrica. Devuelve un conjunto con los elementos que no están en la intersección.**

```python
c1^c2
```

**<span style="color:red">Operador in</span> : en (dentro de). Devuelve un dato de tipo booleano (True, False).**

```python
3 in c1
```

**<span style="color:red">Operador not in</span> : no en (no dentro de). Devuelve un dato de tipo booleano.**

```python
3 not in c1
```

### 3) Algunas funciones predeterminadas

**<span style="color:red">Función len()</span>: Devuelve el número de elementos de un conjunto.**

```python
c3={43, 5, -65, 435}
len(c3)
```

**<span style="color:red">Función sum()</span>: Devuelve la suma de los elementos de un conjunto.**

```python
sum(c3)
```

**<span style="color:red">Función max()</span>: Devuelve el valor máximo del conjunto.**

```python
max(c3)
```

**<span style="color:red">Función min()</span>: Devuelve el valor mínimo del conjunto.**

```python
min(c3)
```

### 4) Métodos de la clase set ###

**<span style="color:red">Método add(objeto)</span>: Añade un objeto al conjunto.**

```python
c4={43, "atgtgc", 57.4, (43,56)}
c4.add("FDGGHERTY")
c4
```

**<span style="color:red">Método remove(objeto)</span>: Elimina un objeto del conjunto.**

```python
c4.remove(43)
c4
```

**<span style="color:red">Método clear</span>: Elimina todos los objetos del conjunto.**

```python
c4.clear()
c4
```
