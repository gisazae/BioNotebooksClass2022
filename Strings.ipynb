# TIPOS DE DATOS EN PYTHON

## CADENAS (strings)

Este tipo de dato es una secuencia de caracteres, y al igual que cualquier tipo de dato en python es un objeto. 
Nos permite ciertos atributos y métodos, especialmente en bioinformatica, trabajar con secuencias de ADN o PROTEINAS. 

### 1)  Para crear una cadena ...

```python
str("atcgtcgtaccgtagtagctacgatgat") #podemos declarar nuestra cadena directamente con el atributo str 
```

```python
"tcgtcgtaccgtagtagctacgtgat" #o simplemente poner el texto entre comillas y python lo reconoce ya como el objeto string. 
```

lo mas recomendado es siempre asignar nuestras cadenas a variables que almacenen la información y resulte mas eficiente el uso de esta dentro del programa. 

```python
#por ejemplo 
dna1="tcgtcgtaccgtagtagctacgtgat" #secuencia 1 asignada o guardada en la variable dna1 
dna2="tcgtcgtaccgtagtagctacgtgatttggccaatagctc" #secuencia 1 asignada o guardada en la variable dna1 
#si quisieramos unir --> concatenar estas dos secuencias, operamos de la siguiente manera:

dna1 + dna2  #el simbolo + nos permite concatenar DATOS DE UN MISMO TIPO. 
```

Otro punto importante dentro de bioinformatica es la necesidad de replicar información para ejecutar experimentos, Para el caso de secuencias especificas como las guardadas en dna1 y dna2 podemos hacer lo siguiente 

```python
#si se desea replicar la informacion de dna1 3 veces, ejecutamos el siguiente comando
dna1="tcgtcgtaccgtagtagctacgtgat"
print(3*dna1)
```

### 2) Operadores de cadenas

UN OPERADOR ES UN SIMBOLO QUE NOS PERMITE REALIZAR OPERACIONES CON UNO O MAS OPERANDOS.

**<span style="color:red">Operador +</span> : Suma. Concatena cadenas.**

```python
"atctgagc"+"atgctgat"
```

**<span style="color:red">Operador *</span> : Multiplicación. Repite cadenas.**

```python
4*"atgctgat"
```

**<span style="color:red">Operador ==</span> : Igual que. Devuelve un dato de tipo booleano (True, False).**

```python
"actg"=="actt"
```

**<span style="color:red">Operador !=</span> : Distinto de. Devuelve un dato de tipo booleano (True, False).**

```python
"actg"!="actt"
```

**<span style="color:red">Operador in</span> : en (dentro de). Devuelve un dato de tipo booleano (True, False).**

```python
"cagta"in"acagtcagtagta"
```

**<span style="color:red">Operador not in</span> : no en (no dentro de). Devuelve un dato de tipo booleano (True, False).**

```python
"cagtt"not in"acagtcagtagta"
```

**<span style="color:red">Operador [ ]</span> : índice. Devuelve el carácter que ocupa la posición indicada dentro de la cadena.**

**En Python se empieza a contar desde el carácter 0.**

```python
"NMSDKLLSETSKMSN"[2]
```

```python
seq="NMSDKLLSETSKMSN"
seq[0]
```

**Si utilizamos números negativos empezamos a contar por el final, a partir del -1.**

```python
seq[-2]
```

**Si queremos obtener una subcadena podemos utilizar intervalos, separados por dos puntos ":".**

```python
seq[2:8]
```

```python
seq[0:-1]
```

```python
seq[0:]
```

```python
seq[:]
```

```python
seq[2:7]
```

```python
seq[:-8]
```

```python
seq[4:30]
```

```python
#seq[30]
```

### 4) Métodos de la clase str ###

**<span style="color:red">Método upper()</span>: Devuelve la cadena en mayúsculas.**

Con los métodos upper() y lower() podemos convertir cadenas en mayúsculas o minúsculas.

```python
newseq=seq.upper()
newseq
```

**<span style="color:red">Método lower()</span>: Devuelve la cadena en minùsculas.**

```python
newseq.lower()
```

**<span style="color:red">Método count(cadena)</span>: Devuelve el número de apariciones de la cadena indicada.**

```python
seq1="actgcgatgatgcaaatttt"
seq1.count("at")
```

**<span style="color:red">Método ljust(int)</span>: Devuelve una cadena como la original pero justificada a la izquierda y con el ancho que le indicamos.**

```python
seq.ljust(20)
```

**<span style="color:red">Método rjust(int)</span>: Devuelve una cadena como la original pero justificada a la derecha y con el ancho que le indicamos.**

```python
seq.rjust(20)
```

**<span style="color:red">Método center(int)</span>: Devuelve una cadena como la original pero centrada y con el ancho que le indicamos.**

```python
seq.center(20)
```

**Podemos indicar el carácter de relleno. Si no indicamos nada, se rellena con espacios por defecto.**

```python
seq.center(20, "*")
```

**<span style="color:red">Método find(cadena)</span>: Devuelve el índice en el que aparece la cadena indicada.**

```python
seq1="actgcgatgatgcaaatttt"
seq1.find("tgcg")
```

**Si no la encuentra devuelve -1.**

```python
seq1="actgcgatgatgcaaatttt"
seq1.find("acgc")
```

**<span style="color:red">Método replace(cadena a, cadena b)</span>: Devuelve una cadena con la cadena a sustituida por la cadena b.**

```python
seq1.replace("a", "t")
```

**Si queremos sustituir solo hasta la segunda "a"**

```python
seq1.replace("a", "t", 2)
```

**<span style="color:red">Método split(cadena)</span>: Devuelve una lista con las cadenas separadas por un saparador que podemos especificar. Si no se especifica se entiende se trata del espacio**

```python
seq2="atagataccagat atgtacagatggtg tgctgatgctgatgc tgtcgattaacgta"
seq2.split()
```

```python
seq2="atagataccagat-atgtacagatggtg-tgctgatgctgatgc-tgtcgattaacgta"
seq2.split("-")
```

**<span style="color:red">Método strip()</span>: Elimina caracteres a la derecha e izquierda de una cadena, devolviendo la nueva cadena. Si no le pasamos ningún parámetro elimina los espacios en blanco.**

```python
seq3="   actgctgca   "
seq3.strip()
```

```python
seq3="+++++actgctgca+++++"
seq3.strip("+")
```

**Del mismo modo rstrip() solo elimina a la derecha de la cadena y lstrip() solo a la izquierda.**

```python
seq3="+++++actgctgca+++++"
seq3.lstrip("+")
```

**<span style="color:red">Método join(iterable)</span>: Toma como argumento una cadena o una lista de cadenas y retorna una cadena resultante separada por la cadena sobre la que se llama al método.**

```python
cad="actg"
"+".join(cad)
```
